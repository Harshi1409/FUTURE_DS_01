# Superstore Sales Analysis
This project analyzes sales data to identify regional performance and insights. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Sales'].sum()
```

```python
df.groupby('Region')['Sales'].sum()
```

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head()
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

## Sales Analysis Insights
- West region has the highest sales.
- South region has the lowest sales.
- East region is performing well.
- Sales distribution is uneven.
